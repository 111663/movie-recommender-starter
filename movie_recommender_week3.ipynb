```python
# import dependencies
# Since pandas and numpy are used a lot, we will import them under short names
# Therefore, we can call pd whenever we use pandas instead of typing out the whole thing
import pandas as pd
import numpy as np

# Load in the data using pandas
df = pd.read_csv('movies_metadata.csv')
```

A great thing about jupyter notebooks is that you can print out variables by just typing their name. Instead of having to type print(df), we can just do df.

```python
df
```

Your output should look like a table. Pandas took in the csv file and transformed it into a dataframe, which can be indexed and transformed easily. As you can see, each row is a movie and each column is a feature that it can have.

Now lets look at the easiest feature, binary ones. These are feature that are either True or False. The adult feature is a binary feature. How would you guys encode it? Think about how Python handles True or False. 1 is equal to True and 0 is equal to False. Lets do that for our features. This is the code to do it.

```python
df_encoded = df.copy() # Copy the df so we preserve the original one
df_encoded['adult'] = df_encoded['adult'].map({'True': 1, 'False': 0})
```

If you make the changes, you will see the values represented as floats.

```python
df_encoded
```

The next simplest are the numerical features. But remember, they often require extra processing to work. Let look at the vote_count feature as an example. To look at a feature use this function.

```python
df['vote_count']
```

Not really useful right? There's no way for you to actually use the information because there are too many rows to look at the actual information. This is where numpy comes in handy. We can look at stats of a row instead the row itself. Let's use these numpy commands

```python
vote_count = df['vote_count']
print(f'{vote_count.max()=}')
print(f'{vote_count.min()=}')
print(f'{vote_count.mean()=}'),
print(f'{vote_count.median()=}'),
print(f'{vote_count.std()=}'),
print(vote_count.quantile([0.25, 0.5, 0.75, 0.95])) # Percentiles for the data
```

This looks more useful right. Now you can learn more about what the data is like. We can see that the max is really high, but the mean is a lot lower. The quick way to adjust these is by using the MinMaxScaler using sklearn. Sklearn is library for machine learning.

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X_normalized = scaler.fit_transform(vote_count.to_numpy().reshape(-1, 1))
```

```python
df_encoded['vote_count'] = X_normalized
```
